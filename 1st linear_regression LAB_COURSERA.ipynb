# Importing the Numpy



```python
import numpy as np
```

Creating our input and output




*   The input represents the SQFT^2 of a house

*   The output represents the prince of it





```python
train_x = np.array([1.0,2.0])
train_y = np.array([300,500])
```

Counting how many training examples we can have in our array of inputs

```python
m = train_x.shape[0]
print(f'Number of training examples', m)
```

Training example x_i, y_i



*   As we can see here x^0 and y^0 represents the first set of inputs and outputs



```python
i = 0
for i in range(len(train_x)):
  print(f"(x^({i}), y^({i})) = ({train_x[i]}, {train_y[i]})")
```

Let's plot a scatter graph to see how it really is

```python
import matplotlib
import matplotlib.pyplot as plt

plt.scatter(train_x, train_y, marker='x')

plt.title("Housing Prices")

plt.ylabel('Price (in 1000s of dollars)')

plt.xlabel('Size (1000 sqft)')
plt.show()
```

the model function for linear regression (which is a function that maps from x to y) is represented as

ùëìùë§,ùëè(ùë•(ùëñ))=ùë§ùë•(ùëñ)+ùëè



w represents the coefficient indicating the price per square foot, and
b could represent additional costs or the base price.



```python
w = 200
b = 100
```

Creating the function that will make the linear regression

```python
def compute_model_output(x,w,b):
  m = train_x.shape[0]
  f_wb = np.zeros(m)

  for i in range(len(x)):
    f_wb[i] = w * x[i]+b

  return f_wb
```

Trying the function

```python
tmp_f_wb = compute_model_output(train_x, w, b)
#ploting the prediction
plt.plot(train_x, tmp_f_wb, c='b', label='Our prediction')
plt.scatter(train_x, train_y, c='r', marker='x', label='Real price')
plt.ylabel('Price $1000s')
plt.xlabel('Size (1000) sqft')
plt.legend()

plt.show()
```
